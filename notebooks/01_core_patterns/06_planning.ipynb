# Planning

Create and execute a plan for multi-step goals.


## Core ideas

- Task decomposition: convert a goal into ordered, observable steps.
- Plan representation: steps need status, dependencies, owner, and completion criteria.
- Execution monitoring: compare actual observations against the plan.
- Replanning: revise when tools fail, context changes, or a step reveals new constraints.
- Granularity: steps that are too broad are hard to verify; steps that are too small waste overhead.


## Common failure modes

- Plans that are too vague to execute.
- Refusing to replan after observations change.
- Optimizing for plan elegance instead of goal completion.


## Framework implementation

Planning is where the course first introduces Deep Agents. Its harness already provides planning, context management, files, and delegation; use it when the task is genuinely open-ended.


```python
# Import the dependencies used by this example.
import os
from deepagents import create_deep_agent
from langgraph.checkpoint.memory import InMemorySaver

if not os.getenv("OPENAI_API_KEY"):
    print("Skipped: set OPENAI_API_KEY to run the Deep Agent.")
else:
    # Configure the framework object; this line prepares it but may not execute it yet.
    agent = create_deep_agent(
        model="openai:gpt-5.6-sol",
        system_prompt=(
            "Create a short plan before acting. Track progress in the todo tool. "
            "Use subagents only for independent research that would clutter the main context."
        ),
        checkpointer=InMemorySaver(),
    )
    config = {"configurable": {"thread_id": "planning-lesson-1"}}
    # Execute the configured model or workflow with the input below.
    result = agent.invoke(
        {"messages": [{"role": "user", "content": "Plan a regression-test rollout for an agent."}]},
        config=config,
    )
    print(result["messages"][-1].content)
```

## Offline mechanics

This version runs without credentials and exposes the control flow directly.


```python
plan = [
    {"step": "collect requirements", "status": "pending"},
    {"step": "choose tools", "status": "pending"},
    {"step": "run solution", "status": "pending"},
    {"step": "verify output", "status": "pending"},
]

# Define the data shape and small operations before running them.
def complete_next(plan):
    for item in plan:
        if item["status"] == "pending":
            item["status"] = "done"
            break
    return plan

complete_next(plan)
complete_next(plan)
plan
```
